```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import copy
```

```python
data1 = pd.read_csv("data1.txt", delimiter=" ", index_col="lp.")
DL = 1 #niedokładność pomiaru - 1mm
Dx_p = DL
Dx_l = DL
d1 = 1/1000 #stała siatki dyfrakcyjnej 1/mm
data1.head()
```

```python
x = (data1['x_p'] + data1['x_l']) / 2 #średnia odległość
data1['x'] = x
data1.head()
```

```python
# def u_Ax(x_p,x_l,x):
#     s = 0
#     for xp,xl in zip(data1['x_p'],data1['x_l']):
#         s += 

def u_B(Deltas):
    uB = 0
    for D in Deltas:
        uB += D**2 / 3.
    uB = uB ** (1./2)
    return uB 
```

```python
uL = u_B([DL])
ux_p = u_B([Dx_p])
ux_l = u_B([Dx_l])
# data1['uL'] = uL
# data1['ux_p'] = ux_p
# data1['ux_l'] = ux_l
# data1[["L",'uL','x_p','ux_p','x_l','ux_l','x']]
```

```python
u_Ax = (((data1['x_l']-data1['x'])**2 + (data1['x_p']-data1['x'])**2) / (2*1))**(1./2)
u_Bx = ux_p
ux = (u_Ax**2 + u_Bx**2)**(1./2)
data1['ux'] = ux
data1
```

```python
sin_theta = data1['x'] / (data1['x']**2 + data1['L']**2)**(1./2)
data1['sin_theta'] = sin_theta

u_Csin_theta = ((data1['L'] * data1['x'] / (data1['L']**2 + data1['x']**2)**(3./2))**2 * uL**2 +
                + (data1['L']**2 / (data1['L']**2 + data1['x']**2)**(3./2))**2 * data1['ux']**2 )**(1./2)
data1['u_Csin_theta'] = u_Csin_theta
```

```python
lambda_i = d1 * data1['sin_theta'] / 1
data1['lambda_i'] = lambda_i
lambda_mean = data1['lambda_i'].mean()

def u_A(X):
    m = X.mean()
    s = 0
    l = X.size
    for x in X:
        s += (x-m)**2
    return (s/l/(l-1))**(1./2)


#print(data1['lambda_i'].mean())
u_Alambda = u_A(data1['lambda_i'])
```

```python
def round_up(n, decimals=0): 
    multiplier = 10 ** decimals 
    return math.ceil(n * multiplier) / multiplier

wynik1 = data1.copy()

wynik1['ux'] = data1['ux'].map(lambda x : round_up(x,2))
wynik1['u_Csin_theta'] = data1['u_Csin_theta'].map(lambda x : round_up(x,4))
wynik1['sin_theta'] = data1['sin_theta'].map(lambda x : round_up(x,4))

wynik1.head()
```

```python
def round_up(n, decimals=0): 
    multiplier = 10 ** decimals 
    return math.ceil(n * multiplier) / multiplier

def how_many_digits(u):
    i = 0
    while (u <= 10):
        u *= 10
        i += 1
    while (100 < u):
        u *= 0.1
        i -= 1
    return i

def roud2significant(X1,X2):
    try:
        Y1 = X1.copy()
        Y2 = X2.copy()
        for i in range(1,len(Y1)+1):
            d = how_many_digits(Y2[i])
            Y1[i] = round(Y1[i],d)
            Y2[i] = round_up(Y2[i],d)
        return Y1, Y2
    except:
        Y1 = X1.copy()
        y2 = copy.deepcopy(X2)
        for i in range(1,len(Y1)+1):
            d = how_many_digits(y2)
            Y1[i] = round(Y1[i],d)
        y2 = round_up(y2,d)
        return Y1

wynik1 = data1.copy()
wynik2 = data2.copy()


wynik1['lambda_i'] = roud2significant(data1['lambda_i'],u_Alambda)
wynik1['lambda_i'] *= 1e6

wynik1['x'], wynik1['ux'] = roud2significant(data1['x'],data1['ux'])
wynik1['sin_theta'], wynik1['u_Csin_theta'] = roud2significant(data1['sin_theta'],data1['u_Csin_theta'])

wynik2['x'], wynik2['ux'] = roud2significant(data2['x'],data2['ux'])
wynik2['sin_theta'], wynik2['u_Csin_theta'] = roud2significant(data2['sin_theta'],data2['u_Csin_theta'])


data1['lambda_i'][1]
u_Alambda
```
